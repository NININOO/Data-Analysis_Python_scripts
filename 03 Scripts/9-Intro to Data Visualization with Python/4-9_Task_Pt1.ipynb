# 03. Data Consistency Checks

### MIXED DATA TYPES

```python
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
customers['first_name'].value_counts(dropna = False)
```

```python
#find NaN value
print(customers['first_name'].isna().sum())
```

Column "first_name" has mixed data because of the missing NaN values.

### MISSING VALUES

```python
customers.isnull().sum()
```

```python
customers.loc[customers['first_name'].isna()]
```

Even though information about the first name is missing, data related to each row is still valuble. So we will keep the data, but change the name to Unknown and change the data type to string.

```python
# Impute the NaN value name with "Unknown"
customers['first_name'].fillna('Unknown', inplace=True)
```

```python
# Fix mixed data type in the "first_name" column to string 
customers['first_name'] = customers['first_name'].astype('str')
```
